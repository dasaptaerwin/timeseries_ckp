```r
# LOAD LIBRARY (LOAD EVERYTIME YOU OPEN THIS CODE)
library('lubridate')
library('tidyverse')
library('forecast')
library('tseries')
library('gridExtra')
library('ggfortify')
```

```r
# open and load data
data<-read.csv('data.csv')
```

```r
head(data)
```

```r
# CURUG PANGANTEN DATA

# TDS data

# Converting time to time series object OK
data$combine <- as.POSIXct(paste(data$DATE, data$TIME_WIB), format="%Y-%m-%d %H:%M") # combining date and time into a column
head(data$combine)
```

```r
# Converting TDS_PPM TEMP_RIVER_C and TEMP_AIR_C to time series data
data$ts.tds <- ts(data$TDS_PPM) # converting number to time series class
data$ts.tempriver <- ts(data$TEMP_RIVER_C)
data$ts.tempair <- ts(data$TEMP_AIR_C)
class(data$ts.tds)  # checking time series class
class(data$ts.tempriver)
class(data$ts.tempair)
```

```r
# combine ts objects
#install.packages('xts')
library('xts')
```

```r
ts.tds <- ts.union(data$combine, data$ts.tds)
head(ts.tds)
```

```r
# Plotting time series data
ts.plot(data$ts.tds, xlab="Bulan",ylab="Total Dissolved Solids (ppm)")
```

```r
ts.plot(data$ts.tempriver, xlab="Bulan",ylab="Temp air sungai (Deg Celcius)")
```

```r
ts.plot(data$ts.tempair, xlab="Bulan",ylab="Temp udara (Deg Celcius)")
```

```r
# fitting time series model
# fit <- stl(data$ts.tds, s.window=5) # not working
fit.tds <- ma(data$ts.tds, order=7) # using moving average (ma) 
fit.tempriver <- ma(data$ts.tempriver, order=7) # using moving average (ma) 
fit.tempair <- ma(data$ts.tempair, order=7) # using moving average (ma) 
```

```r
fit
```

```r
# plotting
plot(data$ts.tds, ylab="TDS (ppm)", xlab="Bulan", col="gray", 
     main="Plot time series TDS")
lines(fit.tds, col="blue", lwd=2)
```

```r
plot(data$ts.tempriver, ylab="Temp air sungai (deg Celcius)", xlab="Bulan", col="gray", 
     main="Plot time series temp air")
lines(fit.tempriver, col="blue", lwd=2)
```

```r
plot(data$ts.tempair, ylab="Temp udara (deg Celcius)", xlab="Bulan", col="gray", 
     main="Plot time series temp udara")
lines(fit.tempair, col="blue", lwd=2)
```

```r
plot(fit.tempair, ylab="Temp (deg Celcius)", xlab="Bulan", col="blue", 
     main="Plot time series Temp air sungai dan udara")
lines(fit.tds, col="red", lwd=2)
```
